### Encode categorical features

```python
def col_to_dummies(df, col):
    return pd.concat(
        [data, pd.get_dummies(data[col], prefix=col, drop_first=True)], 
        axis=1
    ).drop(col, axis=1)
```

```python
data['FireplaceQu'] = data['FireplaceQu'].map({
    np.nan: 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5
})
```

```python
data['HeatingQC'] = data['HeatingQC'].map({
    'Po':0, 'Fa': 1, 'TA': 2, 'Gd': 3, 'Ex': 4
})
```

```python
data['ExterQual'] = data['ExterQual'].map({
    'Po':0, 'Fa': 1, 'TA': 2, 'Gd': 3, 'Ex': 4
})
```

```python
data = col_to_dummies(data, 'Foundation')
```

```python
data['FireplaceGrade'] = data['Fireplaces'] * data['FireplaceQu']
```

```python
data = data.drop(['FireplaceQu', 'Fireplaces', 'YearRemodAdd'], axis=1)
```

# Baseline model
Predict based on overall quality, average price of a mark

Evaluate based on MAE & RMSE

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('SalePrice', axis=1).copy()
```

```python
y = data['SalePrice'].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=101)
```

```python
baseline_preds = data[data.index.isin(X_train.index.tolist())].groupby('OverallQual').mean()[['SalePrice']]
```

```python
baseline_preds = {row.name: row.values[0] for _, row in baseline_preds.iterrows()}
```

```python
y_pred = X_test['OverallQual'].map(baseline_preds)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
def evaluate(y_test, y_pred):
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    print(f'''MAE:\t{mae}\nRMSE:\t{rmse}''')
    plt.figure(figsize=(5,5))
    plt.scatter(y_test, y_pred, alpha=0.4)
    plt.plot(*2*[np.arange(0,500000)], color='r', ls='--')
    plt.xlabel('y true')
    plt.ylabel('y pred')
    plt.ylim(0, 500000)
    plt.xlim(0, 500000)
    plt.show()
```

```python
# Baseline
evaluate(y_test, y_pred)
```

# 1: Linear regression

```python
def eval_model(model, X_train, X_test, y_train, y_test, transform_y=None):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    if transform_y:
        y_pred = transform_y(y_pred)
    evaluate(y_test, y_pred)
    return model
```

## 1.0. No preprocessing

```python
from statsmodels.formula.api import ols
```

```python
model = ols(
    formula=f"SalePrice ~ {' + '.join(data.columns.drop('SalePrice'))}", 
    data=data
).fit()
```

```python
# model.summary()
```

```python
y_pred = model.predict(X_test)
```

```python
evaluate(y_test, y_pred)
```

## 1.1. log-transform for targets

```python
model = ols(
    formula=f"np.log(SalePrice) ~ {' + '.join(data.columns.drop('SalePrice'))}", 
    data=data
).fit()
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.exp(y_pred)
```

```python
evaluate(y_test, y_pred)
```

# 2. Baseline - Random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=5000, min_samples_split=5, max_features=0.5, random_state=42)
```

```python
model = eval_model(model, X_train, X_test, y_train, y_test)
```

```python
def rf_feat_importance(model, data, ascending=False):
    return pd.DataFrame({'cols':data.columns.drop('SalePrice'), 'imp':model.feature_importances_}).sort_values('imp', ascending=ascending)
```

```python
sns.barplot(data=rf_feat_importance(model, data), x='imp', y='cols')
```

# 3. Surpass the baseline - Boosted trees

```python
import lightgbm as lgb
```

```python
model = lgb.LGBMRegressor(random_state=42)
model = eval_model(model, X_train, X_test, y_train, y_test)
```

```python
model = lgb.LGBMRegressor(max_depth=-5, random_state=42)
eval_model(model, X_train, X_test, np.log(y_train), y_test, transform_y=np.exp)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = lgb.LGBMRegressor(random_state=42)
```

```python
param_grid = {
    "max_depth" : [-1, 1, 4, 8, 16, 32, 64, 128],
    "num_leaves" : [31, 1, 8, 16, 64],
    "feature_fraction": [1.0, 0.5, 0.8], 
    "reg_alpha": [0, 0.3, 0.9],
    "reg_lambda": [0, 0.3, 0.9],
    "early_stopping_round": [0, 1, 2, 5, 10],
}
```

```python
grid = GridSearchCV(
    estimator=model,
    scoring='neg_mean_squared_error',
    param_grid=param_grid,
    cv=10,
    verbose=-1,
)
```

```python
%%capture
grid.fit(
    X_train, 
    np.log(y_train), 
    eval_set=[(X_train, np.log(y_train)), (X_test, np.log(y_test))],
    eval_names=['train', 'valid'], 
    eval_metric='rmse', 
#     verbose=0
)
```

```python
grid.best_params_
```

```python
y_pred = np.exp(grid.predict(X_test))
```

```python
evaluate(y_test, y_pred)
```
